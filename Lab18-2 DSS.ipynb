## Aufgabe: Datei einlesen und visualisieren

```python
import pandas as pd

name = "Natalie"
gender = "F"
state = "CA"

df = pd.read_csv("../data/names.csv")
len(df)
```

```python
name_df = df[df["Name"] == name]
state_df = name_df[name_df["State"] == state]
f_df = state_df[state_df["Gender"] == gender]

# min_year = f_df["Year"].min()
# max_year = f_df["Year"].max()

# ~twice as slow
m_df = df[(df["Name"] == "Max") & (df["State"] == state) & (df["Gender"] == "M")]

print(f_df.head())
```

```python
# %matplotlib notebook
# %matplotlib tk
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(f_df["Year"], f_df["Count"], color="#FF69B4", label="\"Natalie\" count")
plt.plot(m_df["Year"], m_df["Count"], color="#1155FF", label="\"Max\" count")
plt.legend()
plt.show()
```

```python
f_sum = f_df["Count"].sum()
m_sum = m_df["Count"].sum()
plt.pie(
    [f_sum, m_sum], 
    labels = ["{0} {1} times".format(name, f_sum), "Max {0} times".format(m_sum)], 
    colors = ["#FF69B4", "#1155FF"],
    shadow = True
)
plt.show()
```
